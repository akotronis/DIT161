#### Data Column Info 2

- **Departures**: Data on ***all flights*** between US gateways and non-US gateways, irrespective of origin and destination.
Each **observation** provides information on a specific airline for a pair of airports, one in the US and the other outside. Three main columns record the number of flights: Scheduled, Charter, and Total.
- **Passengers**: Data on the ***total number of passengers*** for each month and year between a pair of airports, as serviced by a particular airline.

U.S. International Air Passenger and Freight data are confidential for a period of 6 months, after which it can be released. As a result, quarterly reports and the year to date/calendar year raw data files available here will always lag by two quarters

- **Airport Codes** [Data Column Description](https://ourairports.com/help/data-dictionary.html) | [Other usefull info](https://ourairports.com/help/#legend)

```python
import calendar

calendar.setfirstweekday(calendar.MONDAY)
days_names = list(calendar.day_name)
print(days_names)
month_names = [m for m in list(calendar.month_name) if m.strip()]
print(month_names)
```

```python
time_dim = pd.DataFrame(date_values, columns=['date'])
time_dim[['day', 'month', 'year']] = pd.DataFrame(
    zip(time_dim['date'].dt.day, time_dim['date'].dt.month, time_dim['date'].dt.year),
    columns=['day', 'month', 'year']
)

weekdays_df = pd.DataFrame.from_dict({'weekday':days_names})
months_df = pd.DataFrame.from_dict({'month':month_names})
months_df.index = months_df.index + 1
seasons_df = pd.DataFrame.from_dict({'season':['Winter', 'Spring', 'Summer', 'Autumn']})

week_days_swapped = pd.Series(weekdays_df['weekday'].index.values, index=weekdays_df['weekday'])
months_swapped = pd.Series(months_df['month'].index.values, index=months_df['month'])
time_dim['day_of_week_key'] = time_dim['date'].dt.strftime('%A').map(week_days_swapped)
time_dim['season_key'] = time_dim['month'] % 12 // 3
time_dim['day_of_year'] = time_dim['date'].dt.strftime('%j').astype(int)
time_dim['week_of_year'] = time_dim['date'].dt.strftime('%W').astype(int) # Monday as the first day of the week

time_dim['date'] = time_dim.index
time_dim = time_dim.rename(columns={'date':'id'})
dspl(time_dim.head(2))
```

Define function to clean unwanted text from entries

```python
def remove_unwanted_stuff(txt):
    if pd.isnull(txt):
        return np.nan
    else:
        txt = txt.replace('*', '').replace('—', '').replace('n/a', '').replace('N/A', '')
        txt = ' '.join(txt.split())
        if not txt:
            return np.nan
        else:
            if not ('[' in txt and ']' in txt):
                return txt
            else:
                start = txt.find('[')
                end = txt.find(']')
                repl = txt[start:end + 1]
                txt = txt.replace(repl, '')
                return remove_unwanted_stuff(txt)

tests = ['cdsd', 'cd[32]sd', 'cd[32]sd[f]', np.nan]
for test in tests:
    print(remove_unwanted_stuff(test))
```

We define a function to modify columns according to the model (Classification - Regression)

From **Activity Period** we remove **year** as we want to predict counts for future time (See [here](https://towardsdatascience.com/make-your-machine-learning-model-work-better-with-datetime-features-eb21de397fe8) and [here](https://towardsdatascience.com/machine-learning-with-datetime-feature-engineering-predicting-healthcare-appointment-no-shows-5e4ca3a85f96))

```python
from dython.nominal import associations #, identify_nominal_columns

def prepare_data(df, month_to_trig=False, drop_columns=None, print_corrs=True, model='clf'):
    '''model = clf (classification) or rgr (regression)
       If month_to_trig=True we make two columns from month, converted with sin/cosine
    '''
    
    # Handle month (drop Activity Period in either case)
    if 'Month' not in drop_columns:
        if month_to_trig:
            df.insert(0, 'Month_sin', np.sin(2 * np.pi * (df['Activity Period'] % (df['Activity Period'] // 100)) / 12.))
            df.insert(0, 'Month_cos', np.cos(2 * np.pi * (df['Activity Period'] % (df['Activity Period'] // 100)) / 12.))
        else:
            df.insert(0, 'Month', df['Activity Period'] % (df['Activity Period'] // 100))
    
    # Drop other cols
    df = df.drop(['Activity Period'], axis='columns')
    for c in drop_columns:
        if c in df.columns:
            df = df.drop(c, axis='columns')

    # print correlations
    if print_corrs:
        print(100 * '=')
        print(f" FEATURE CORELLATIONS ".center(100, '='))
        print(100 * '=')
        associations(df.iloc[:,:-1], figsize=(8,8))['corr'].dropna(axis=1, how='all').dropna(axis=0, how='all')
    
    if model == 'clf':
        bins = [0, 3_500, 7_000, 11_500, 30_000, 700_000]
        labels = ['E', 'D', 'C', 'B', 'A']
        df['classes'] = pd.cut(df['Passenger Count'], bins=bins, labels=labels)
        predictors = df.loc[:,[c for c in df.columns if c not in ['classes', 'Passenger Count']]]
        target = df['classes']
    else:
        predictors = df.loc[:,[c for c in df.columns if c != 'Passenger Count']]
        target = df['Passenger Count'].astype(float)
    return predictors, target
```

Define Function to calculate and plot feature importances

```python
def calc_feature_importance(features, feature_importances, plot='pie'):
    '''plot = pie, bar or None'''
    columns = [('_'.join(c.split('_')[:-1]), c.split('_')[-1]) for c in features]
    m_index = pd.MultiIndex.from_tuples(columns, names=['Features', ''])
    df = pd.DataFrame(data=feature_importances, index=m_index, columns=['Importances']).groupby('Features').sum()
    df = (100 * df / df.sum()).round(3)
    df = df.sort_values('Importances', ascending=False)
    if plot == 'pie':
        _ = df['Importances'].plot(kind='pie', legend=None, y='Importances', xlabel='', ylabel='', autopct='%.2f')
    elif plot == 'bar':
        _ = df['Importances'].plot(kind='bar', legend=None, ylabel='Importances (%)', xlabel='')
    return df
```

Define a function that makes a pipeline to hot encode categorical features and scale numeric. Returns the pipeline

```python
from sklearn.compose import ColumnTransformer #, make_column_transformer
from sklearn.feature_selection import f_classif, SelectKBest #, chi2
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.naive_bayes import GaussianNB

def data_pipeline(model=None, numeric_features=None, categorical_features=None):
    # https://scikit-learn.org/stable/auto_examples/compose/plot_column_transformer_mixed_types.html
    # Scale numerical features
    numeric_transformer = Pipeline(
        steps=[("scaler", StandardScaler())]
    )
    # One Hot Encode categorical features
    categorical_transformer = OneHotEncoder(handle_unknown="ignore")
    preprocessor = ColumnTransformer(
        transformers=[
            ("num", numeric_transformer, numeric_features),
            ("cat", categorical_transformer, categorical_features),
        ],
        remainder='passthrough', # leave rest of features as they are
        sparse_threshold=0       # Always return dense data (spase won't work with NaiveB for example)
    )
    
    steps=[("preprocessor", preprocessor)]
    if model is not None:
        steps.append(("model", model))
    pipeline = Pipeline(
        steps=steps
    )
    return pipeline
```

Find best model: **Hyperparameter tuning** and **cross validation** with `GridSearchCV`

We use **train** data on `GridSearchCV` and perform **5-fold** cross validation along with parameter tuning to find best model

Define a function to do this that accepts as input the pipeline with a selected classifier, the corresponding parameters and a metric to evaluate upon

```python
def fine_tune(ppl, param_grid, metric, X_train, y_train, n_splits=5, model='clf'):
    '''model=clf (Classification) or rgr (Regression)'''
    from sklearn.model_selection import GridSearchCV
    from sklearn.model_selection import StratifiedKFold
    
    # Classification: By default uses stratified sampling for binary/multiclass
    # https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
    # but anyway we useStratifiedKFold
    splits = StratifiedKFold(n_splits=n_splits) if model == 'clf' else n_splits
    grid = GridSearchCV(
        ppl,
        param_grid,
        cv=splits,
        n_jobs=-1,
        scoring=metric,
        return_train_score=True,
    )
    grid.fit(X_train, y_train)
    return grid
```

Define a function to plot and return results of hyperparameter tuning

```python
def fine_tune_report(result_grid, plot_param=None):
    '''Accepts the gridsearch result and a parameter to plot for train-test score
        Return tuple with best estimator and results as dataframe
    '''
    df = pd.DataFrame(result_grid.cv_results_)
    df_return = df.copy()
    if plot_param:
        plot_param_title = ' '.join([x.title() for x in plot_param.split('_')])
        df = df[[f"param_model__{plot_param}" , 'mean_train_score', 'mean_test_score']]
        df = df.rename(columns={'mean_test_score':'Test','mean_train_score':'Train', f"param_model__{plot_param}":plot_param_title})
        df = df.groupby(plot_param_title).mean()
        _ = df.plot(kind='line', grid=True, ylabel='Mean Score')
    return {'best_estimator':result_grid.best_estimator_['model'], 'results_df':df_return}
```

Define a classification report function with **roc_auc** added

```python
def clf_report(txt, roc_auc_scores=None, roc_auc_mavg=None, roc_auc_wavg=None):
    txt = txt.strip().replace('\n\n', '\n')
    txt = [ln.strip() for ln in txt.split('\n')]
    lines = []
    for ln in txt:
        if ln.startswith('accuracy'):
            line = ln.split()
            line.insert(1,'')
            line.insert(1,'')
        elif ln.startswith('precision'):
            line = ln.split()
            line.insert(0,'classes/agg')
        elif 'avg' in ln:
            line = ln.replace(' avg','-avg').split()
        else:
            line = ln.split()
        lines.append(line)
    
    df = pd.DataFrame(data=lines[1:], columns=lines[0]).set_index('classes/agg')
    if all([roc_auc_scores, roc_auc_mavg, roc_auc_wavg]):
        roc_auc_col = roc_auc_scores + [df.loc['accuracy', 'support']] + [roc_auc_mavg, roc_auc_wavg]
        df['roc_auc'] = roc_auc_col
    return df
```

Define function to calculate **roc auc scores**

```python
def roc_auc_scores(ppl, X, y, classes):
    y_pred_probs = ppl.predict_proba(X)
    roc_aucs = []
    for cls in classes:
        idx = np.where(ppl.classes_ == cls)[0][0]
        roc_aucs.append(roc_auc_score(y == cls, y_pred_probs[:, idx]))
    roc_auc_mavg = roc_auc_score(y, y_pred_probs, multi_class="ovr", average='macro')
    roc_auc_wavg = roc_auc_score(y, y_pred_probs, multi_class="ovr", average='weighted')
    roc_aucs.extend([roc_auc_mavg, roc_auc_wavg])
    return roc_aucs
```

**Method 1**: IQR

```python
import seaborn as sns

_ = sns.boxplot(x=target)
```

```python
Q1 = target.quantile(0.25)
Q3 = target.quantile(0.75)
IQR = Q3 - Q1
outlier_mask = (target < Q1 - 1.5 * IQR) | (target > Q3 + 1.5 * IQR)

outliers = target.loc[outlier_mask]
rest = target.loc[~outlier_mask]
outl_num = len(outliers)
rest_num = len(rest)
trg_num = len(target)
print(f"All: {trg_num}\nOutliers: {outl_num} ({100 * outl_num/float(trg_num):.2f}%)\nRest: {rest_num} ({100 * rest_num/float(trg_num):.2f}%)")
```

The transformed data approach normal distribution so we can use **zscore** to detect outliers with threshold **3**

```python
m = np.mean(target_trsf)
s = np.std(target_trsf)
zscore = (target_trsf - m)/s
outlier_mask = abs(zscore) > 3

outliers = target.loc[outlier_mask]
rest = target.loc[~outlier_mask]
outl_num = len(outliers)
rest_num = len(rest)
trg_num = len(target)
print(f"All: {trg_num}\nOutliers: {outl_num} ({100 * outl_num/float(trg_num):.2f}%)\nRest: {rest_num} ({100 * rest_num/float(trg_num):.2f}%)")
```

**Method 2**:  `IsolationForest`

```python
# https://www.knime.com/blog/four-techniques-for-outlier-detection
from sklearn.ensemble import IsolationForest

clf = IsolationForest(
    random_state=1,
    max_samples=100,
)
table = target.array.reshape(-1,1)
clf.fit(table)
outlier_mask = pd.Series(clf.predict(table)).map({-1:True, 1:False}).array

outliers = target.loc[outlier_mask]
rest = target.loc[~outlier_mask]
outl_num = len(outliers)
rest_num = len(rest)
trg_num = len(target)
print(f"All: {trg_num}\nOutliers: {outl_num} ({100 * outl_num/float(trg_num):.2f}%)\nRest: {rest_num} ({100 * rest_num/float(trg_num):.2f}%)")
```

**Method 3**:  `DBSCAN`

```python
# https://medium.com/@dilip.voleti/dbscan-algorithm-for-fraud-detection-outlier-detection-in-a-data-set-60a10ad06ea8
# https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(
        eps=200
)
table = target.array.reshape(-1,1)
dbscan.fit(table)

outlier_mask = pd.Series(dbscan.labels_).map(lambda x: True if x == -1 else False).array
outliers = target.loc[outlier_mask]
rest = target.loc[~outlier_mask]
outl_num = len(outliers)
rest_num = len(rest)
trg_num = len(target)
print(f"All: {trg_num}\nOutliers: {outl_num} ({100 * outl_num/float(trg_num):.2f}%)\nRest: {rest_num} ({100 * rest_num/float(trg_num):.2f}%)")
```

**Method 4**:  `LocalOutlierFactor`

```python
# https://medium.com/mlearning-ai/how-to-find-and-remove-outliers-from-a-regression-in-python-449bc9e13101
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(
    contamination=0.01,
    n_neighbors=20,
    novelty=False,
)

table = target.array.reshape(-1,1)
result = lof.fit_predict(table)

outlier_mask = pd.Series(result).map(lambda x: True if x == -1 else False).array
outliers = target.loc[outlier_mask]
rest = target.loc[~outlier_mask]
outl_num = len(outliers)
rest_num = len(rest)
trg_num = len(target)
print(f"All: {trg_num}\nOutliers: {outl_num} ({100 * outl_num/float(trg_num):.2f}%)\nRest: {rest_num} ({100 * rest_num/float(trg_num):.2f}%)")
```
